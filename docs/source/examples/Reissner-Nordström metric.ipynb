# Importing Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
from einsteinpy.symbolic import MetricTensor, ChristoffelSymbols, RiemannCurvatureTensor
from einsteinpy.geodesic import Geodesic
from mpl_toolkits.mplot3d import Axes3D
```

## Defining the metric function


```python
def reissner_nordstrom_metric(mass, charge, coords):
    #show the variables: radical coordinates, theta and phi (for spherical angles) 
    #M - mass of object // Q - elecrical charge of object
    #gtt (radical)- the componet matches to the cooef of dt^2 metric tensor 
    #gθθ (theta)- matches to the coefficient of dθ^2 in metric tensor.
    #gϕϕ (phi)- matches to the coefficient of dϕ^2 in the metric tensor.

    r, theta, phi = coords
    gtt = 1 - 2 * mass / r + charge**2 / r**2
    grr = 1 / gtt
    gTheta = r**2
    gphi = (r**2) * (np.sin(theta)**2)
    
    #return metric tensor components
    return [[gtt, 0, 0, 0],
            [0, grr, 0, 0],
            [0, 0, gTheta, 0],
            [0, 0, 0, gphi]]
```

## Compute Metric Components

```python
#mass and charge
mass = 1.0
charge = 0.5

#spatial coordinates
r_values = np.linspace(1, 10, 100)
theta_values = np.linspace(0, np.pi, 100)
phi_values = np.linspace(0, 2*np.pi, 100)



metric_components = reissner_nordstrom_metric(mass, charge, (r_values, theta_values, phi_values))
```

## visualise the metric

```python
#plotting metric components

plt.plot(r_values, metric_components[1][1], label='$g_{rr}$')
plt.xlabel('r')
plt.ylabel('$g_{rr}$')
plt.title('Reissner-Nordström Metric $g_{rr}$')
plt.legend()
plt.grid(True)
plt.show()
```
